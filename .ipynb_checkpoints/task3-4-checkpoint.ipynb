# Data Clustering Functions

##Â Dimensionality Reduction
Dimensionality reduction techniques allow us to overcome the limitations posed by high-dimensional data and potentially improve clustering performance. The two methods available are:
- **Principal Component Analysis (PCA)**
- **Multi-dimensional Scaling (MDS)**

```python
def perform_dimensionality_reduction(X, method='pca'):
    """
    Perform dimensionality reduction on the data.

    Args:
        X (np.ndarray): The input dataset.
        method (str, optional): The dimensionality reduction method to use. Can be 'pca' or 'mds'. 
            Defaults to 'pca'.

    Returns:
        np.ndarray: The dimensionality-reduced dataset.
    """
    if method == 'pca':
        pca = PCA(n_components=2)
        X_reduced = pca.fit_transform(X)
    elif method == 'mds':
        mds = MDS(n_components=2)
        X_reduced = mds.fit_transform(X)
    else:
        raise ValueError("Invalid dimensionality reduction specified. Please choose 'pca' or 'mds'.")
    return X_reduced
```

## Silhouette scores plot

```python
def plot_silhouette_scores(min_clusters, max_clusters, silhouette_scores):
    """
    Plot the silhouette scores for a given set of clustering parameters.

    Parameters:
        min_clusters (int): The minimum number of clusters.
        max_clusters (int): The maximum number of clusters.
        silhouette_scores (list): List of silhouette scores.

    Returns:
        None
    """
    plt.plot(range(min_clusters, max_clusters+1), silhouette_scores, marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Score vs. Number of Clusters')
    plt.show()
```

## Parameter Optimisation for Clustering

```python
def perform_grid_search(X, param_grid):
    """
    Perform grid search to find the best parameters for clustering.

    Parameters:
        X_reduced (np.ndarray): The dimensionality-reduced dataset.
        min_clusters (int): The minimum number of clusters to consider.
        max_clusters (int): The maximum number of clusters to consider.

    Returns:
        float: The best Silhouette score achieved.
        dict: The best parameters for clustering.
    """
    silhouette_scores = []
    best_score = -1
    best_params = {}

    for params in ParameterGrid(param_grid):
        kmeans = KMeans(random_state=2002, **params)
        labels = kmeans.fit_predict(X)
        silhouette = silhouette_score(X, labels)
        silhouette_scores.append(silhouette)

        if silhouette > best_score:
            best_score = silhouette
            best_params = params
            print(best_params)

    return best_score, best_params, silhouette_scores
```

## Performance Evauation Functions

###  Cluster Label Mapping 

```python
def map_cluster_labels(cluster_labels, y_true):
    """
    Map cluster labels to class labels based on majority voting.

    Parameters:
        cluster_labels (numpy.ndarray): Cluster labels assigned by the clustering algorithm.
        y_true (numpy.ndarray): The true class labels.

    Returns:
        np.ndarray: The mapped class labels corresponding to the cluster labels.
    """
    # Initialise cluster associations
    cluster_assoc = -1 * np.ones(len(np.unique(cluster_labels)))
    
    # Determine the majority class in each cluster
    for i, cluster in enumerate(np.unique(cluster_labels)):
        cluster_samples = y_true[cluster_labels == cluster]
        unique_classes, class_counts = np.unique(cluster_samples, return_counts=True)
        cluster_assoc[i] = unique_classes[np.argmax(class_counts)]
    
    # Map cluster labels to class labels
    mapped_labels = [cluster_assoc[label] for label in cluster_labels]
    return np.array(mapped_labels)
```

### Regression Metric Calculation

```python
def evaluate_regression_performance(y_true, y_pred):
    """
    Evaluate the regression performance using various metrics.

    Parameters:
        y_true (numpy.ndarray): The true regression targets.
        y_pred (numpy.ndarray): The predicted regression targets.

    Returns:
        dict: A dictionary containing the evaluation metrics.
            - 'mean_squared_error': The mean squared error.
            - 'mean_absolute_error': The mean absolute error.
            - 'r2_score': The coefficient of determination (R^2 score).
    """
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    evaluation_metrics = {
        'mean_squared_error': mse,
        'mean_absolute_error': mae,
        'r2_score': r2
    }

    print("Clustering Evaluation Metrics for Label Predictions:")
    for metric, value in evaluation_metrics.items():
        print(metric + ":", value)

    return evaluation_metrics
```

### Classification Metric Calculation

```python
def evaluate_classification_performance(y_true, y_pred):
    """
    Evaluate the classification performance using the mapped labels.

    Parameters:
        y_true (numpy.ndarray): The true class labels.
        y_pred (numpy.ndarray): The predicted class labels.

    Returns:
        dict: A dictionary containing the evaluation metrics.
            - 'accuracy': The accuracy score.
            - 'precision': The precision score (macro-averaged).
            - 'recall': The recall score (macro-averaged).
            - 'f1': The F1-score (macro-averaged).
    """
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    f1 = f1_score(y_true, y_pred, average='macro')

    evaluation_metrics = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
    
    print("Clustering Evaluation Metrics for Label Predictions:")
    for metric, value in evaluation_metrics.items():
        print(metric + ":", value)

    return evaluation_metrics
```

## Main Clustering Pipeline

```python
def cluster_data(X_train, X_test, y_test, model_type, min_clusters=2, max_clusters=10, dr_method='pca'):
    """
    Cluster the data using the K-means algorithm with optional dimensionality reduction.

    Parameters:
        X_train (np.ndarray): The training data.
        X_test (np.ndarray): The testing data.
        min_clusters (int, optional): The minimum number of clusters to consider. Defaults to 2.
        max_clusters (int, optional): The maximum number of clusters to consider. Defaults to 10.
        dr_method (str, optional): The dimensionality reduction method to use. Can be 'pca' or 'mds'. 
            Defaults to 'pca'.

    Raises:
        ValueError: If an invalid dimensionality reduction method is specified.

    Returns:
        float: The best Silhouette score achieved.
        dict: The best parameters for clustering.
    """
    # Remove the last column (target variable) from training/testing data
    X_train = X_train[:, :-1]
    X_test = X_test[:, :-1]
    
    if dr_method is not None:
        # Perform dimensionality reduction on training data
        X_train = perform_dimensionality_reduction(X_train, dr_method)

        # Perform dimensionality reduction on testing data
        X_test = perform_dimensionality_reduction(X_test, dr_method)
    
    # Define parameter grid for grid search
    param_grid = {'n_clusters': range(min_clusters, max_clusters+1)}

    silhouette_scores = []
    best_score = -1
    best_params = {}
    
    # Perform Grid search to find the optimal number of clusters for the training data and plot the results
    if model_type == "regression":
        best_score, best_params, silhouette_scores = perform_grid_search(X_train, param_grid)
        plot_silhouette_scores(min_clusters, max_clusters, silhouette_scores)
    elif model_type == "classification":
        n_labels = len(np.unique(y_test))
        best_params = {'n_clusters': n_labels}
        kmeans = KMeans(random_state=2002, **best_params)
        labels = kmeans.fit_predict(X_train)
        best_score = silhouette_score(X_train, labels)

    # Perform clustering on training data with best parameters
    kmeans = KMeans(random_state=2002, **best_params)
    train_labels = kmeans.fit_predict(X_train)
    train_centroids = kmeans.cluster_centers_

    # Perform clustering on testing data with best parameters
    test_labels = kmeans.predict(X_test)
    
    # Map cluster labels to class labels for testing data
    mapped_test_labels = map_cluster_labels(test_labels, y_test)
    
    if model_type == "regression":
        # Calculate evaluation metrics from predicted labels from clusterig
        evaluate_regression_performance(y_test, mapped_test_labels)
    elif model_type == "classification":
        # Calculate evaluation metrics from predicted labels from clusterig
        evaluate_classification_performance(y_test, mapped_test_labels)

    # Visualise the clusters and centroids on training data
    plt.scatter(X_train[:, 0], X_train[:, 1], c=train_labels)
    plt.scatter(train_centroids[:, 0], train_centroids[:, 1], marker='x', color='red', label='Centroids')
    plt.legend()
    plt.title('Clustering with Best Parameters (Training Data)')
    plt.show()

    # Visualise the clusters on testing data
    plt.scatter(X_test[:, 0], X_test[:, 1], c=test_labels)
    plt.scatter(train_centroids[:, 0], train_centroids[:, 1], marker='x', color='red', label='Centroids')
    plt.legend()
    plt.title('Clustering with Best Parameters (Testing Data)')
    plt.show()

    print("Best Silhouette Score:", best_score)
    print("Best Parameters:", best_params)
```

This explains the superior performance of the SVC and KNN Classifiers over K-Means clustering when used with the star dataset.
