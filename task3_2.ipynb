# Model Training Functions

The following functions play a crucial role in facilitating the training and optimisation of our data mining models. They provide a comprehensive set of tools to streamline the model training process, fine-tune model parameters, and evaluate model performance.

## Split Data into Training and Testing Sets for Modelling

```python
def split_data(data, test_size=0.2, random_state=2002):
    """
    Split the data into training and testing sets, assuming the last column is the target (y) and the columns before it
    are the predictors (X).

    Parameters:
        data (numpy.ndarray): The input numpy.ndarray representing the dataset.
        test_size (float, optional): The proportion of the dataset to include in the testing set.
            Defaults to 0.2.
        random_state (int, optional): The seed used by the random number generator for reproducible results.
            Defaults to 42.

    Returns:
        tuple: A tuple containing the training and testing sets, in the order (X_train, X_test, y_train, y_test).

    """
    X = data[:, :-1]
    y = data[:, -1]

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state)

    return X_train, X_test, y_train, y_test
```

## Plot Regression Model Results

```python
def plot_regression_results(model, X_test, y_test, title):
    """
    Plot the regression results by comparing the true values and the predicted values.

    Parameters:
        model: The trained regression model.
        X_test (numpy.ndarray): The feature values of the test set.
        y_test (numpy.ndarray): The true target values of the test set.
        title (str): The title of the plot.

    Returns:
        None

    """
    # Predict target values using the trained model
    y_pred = model.predict(X_test)

    # Calculate mean squared error and mean absolute error
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    # Create scatter plot of true values vs. predicted values
    plt.scatter(y_test, y_pred)
    plt.xlabel("True Values")
    plt.ylabel("Predictions")
    plt.title(f"{title}\nMSE: {mse:.2f}, MAE: {mae:.2f}")
    
    # Add a diagonal line for reference
    plt.plot([0, np.max(y_test)], [0, np.max(y_test)], 'k--')
    
    # Show the plot
    plt.show()
```

## Optimise Regression Models

```python
def optimise_regression_model(model, params, X_train, y_train, X_test, y_test, display_results=True):
    """
    Optimise a machine learning regression model by performing grid search with cross-validation (GridSearchCV).

    Parameters:
        model: The regression model to be optimised.
        params (dict): The hyperparameters to be explored within the grid search.
        X_train (numpy.ndarray): The feature values of the training set.
        y_train (numpy.ndarray): The true target values of the training set.
        X_test (numpy.ndarray): The feature values of the test set.
        y_test (numpy.ndarray): The true target values of the test set.
        display_results (bool, optional): Whether or not to display the optimisation results. Defaults to True.

    Returns:
        tuple: A tuple containing the best parameters found, the cross-validation results, and the best estimator.

    """
    # Perform grid search with cross-validation
    grid = GridSearchCV(model, params, refit=True, verbose=0, n_jobs=-1, scoring='neg_mean_squared_error')
    grid.fit(X_train, y_train)

    # Print best parameters after tuning
    if display_results:
        display(Markdown("### Best Parameters"))
        print(grid.best_params_)

    # Predict target values using the best estimator
    grid_predictions = grid.predict(X_test)

    # Plot predicted vs true values
    if display_results:
        plot_regression_results(grid, X_test, y_test, 'Regression Results')

    # Extract the results of the grid search
    cv_results = grid.cv_results_
    best_estimator = grid.best_estimator_
    best_params = grid.best_params_

    return best_params, cv_results, best_estimator
```

## Multi-class Precision Recall Curve Plot for Classification Model Results

```python
def plot_precision_recall_curve(model, X_test, y_test):
    """
    Plot an adapted precision-recall curve for a multi-class classification model. 
    A precision recall curve is created for each class present in the dataset and plotted on a single axis.

    Parameters:
        model: The trained classification model object.
        X_test (numpy.ndarray): The feature values of the testing set.
        y_test (numpy.ndarray): The true target values of the testing set.

    Returns:
        None

    """
    # Get predicted probabilities for test data
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)
    else:
        y_prob = np.zeros((len(X_test), len(model.best_estimator_.classes_)))

        # Iterate over each unique target label from the model's training
        for i, label in enumerate(model.best_estimator_.classes_):
            # Compute the decision function scores (confidence of assignment, alternative to prediction probabbilities)
            # for the corresponding class label
            y_prob[:, i] = model.decision_function(X_test)[:, i]


    # Calculate precision and recall for each class
    precision = dict()
    recall = dict()
    thresholds = dict()
    for i, label in enumerate(model.classes_):
        precision[label], recall[label], thresholds[label] = precision_recall_curve(
            y_test == label, y_prob[:, i])

    # Plot the precision-recall curve for each class
    plt.figure(figsize=(8, 6))
    for label in model.classes_:
        plt.plot(recall[label], precision[label], lw=2,
                 label=f"{label} (AP = {np.mean(precision[label]):.2f})")

    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title("Precision-Recall Curve")
    plt.legend(loc="best")
    plt.show()
```

## Confusion Matrix Plot

```python
def plot_confusion_matrix(y_true, y_pred):
    """
    Plot a confusion matrix for classification performance evaluation. This illustrates the number of correct predictions for
    each class label using a heatmap.

    Parameters:
        y_true (numpy.ndarray): The true (actual) class labels.
        y_pred (numpy.ndarray): The predicted class labels.

    Returns:
        None
    """
    cm = confusion_matrix(y_true, y_pred)
    num_classes = len(np.unique(y_true))
    
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    tick_marks = np.arange(num_classes)
    plt.xticks(tick_marks)
    plt.yticks(tick_marks)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    
    for i in range(num_classes):
        for j in range(num_classes):
            plt.text(j, i, str(cm[i, j]), horizontalalignment='center', verticalalignment='center')

    plt.show()
```

## Optimise Classification Model

```python
def optimise_classifier_model(model, params, X_train, y_train, X_test, y_test, display_results=True):
    """
    Optimise a classifier model using grid search.

    Parameters:
        model (object): The classifier model to optimise.
        params (dict): Hyperparameter grid for the model.
        X_train (numpy.ndarray): Training features.
        y_train (numpy.ndarray): Training labels.
        X_test (numpy.ndarray): Test features.
        y_test (numpy.ndarray): Test labels.
        display_results (bool, optional): Whether to display optimisation results. Defaults to True.

    Returns:
        tuple: Tuple containing the best parameters, cross-validation results, and best estimator.

    """
    # Perform grid search
    grid = GridSearchCV(model, params, refit=True, verbose=0, n_jobs=-1)
    grid.fit(X_train, y_train)

    # Print best parameters after tuning
    if display_results:
        display(Markdown("### Best Parameters"))
        print(grid.best_params_)

    # Make predictions on the test set
    grid_predictions = grid.predict(X_test)

    if display_results:
        # Print classification report
        print(classification_report(y_test, grid_predictions))
        plot_confusion_matrix(y_test, grid_predictions)

    # Extract the results of the grid search
    cv_results = grid.cv_results_
    best_estimator = grid.best_estimator_
    best_params = grid.best_params_

    if display_results:
        # Plot precision-recall curve
        plot_precision_recall_curve(grid, X_test, y_test)

    return best_params, cv_results, best_estimator
```

## Dataset Random Sampling

```python
def create_proportional_subset(array, proportion):
    """
    Create a proportional subset of a numpy ndarray, randomly.

    Parameters:
        array (numpy.ndarray): The input numpy ndarray (a dataset in this case).
        proportion (float): Proportion of rows to include in the subset (between 0 and 1).

    Returns:
        numpy.ndarray: Subset of the input array.

    Raises:
        ValueError: If the proportion is outside the valid range (between 0 and 1).

    """
    # Check if proportion is within valid range
    if proportion <= 0 or proportion > 1:
        raise ValueError("the proportion should be between 0 and 1 (exclusive).")

    # Calculate the number of rows to include in the subset
    nrows = int(np.round(len(array) * proportion))

    # Create a random sample of row indices
    indices = np.random.choice(len(array), nrows, replace=False)

    # Get the rows corresponding to the selected indices
    subset = array[indices]

    return subset
```

## Grid Search Results Plotting

### Single Parameter Results Plot

```python
def plot_grid_search_one_param(cv_results, grid_param, param_name, ax):
    """
    Plot the grid search results for a single parameter. The average score (Y) against the paramater range (X).

    Parameters:
        cv_results (dict): The dictionary of cross-validation results from grid search.
        grid_param (numpy.ndarray): The values of the parameter being optimised.
        param_name (str): The name of the parameter being optimised.
        ax (AxesSubplot): The matplotlib Axes object to draw the plot. Required to combine multiple subplots. 

    Returns:
        None

    """
    # Extract the relevant keys from cv_results
    scores_mean = cv_results['mean_test_score']
    scores_sd = cv_results['std_test_score']
    param_values = cv_results['param_' + param_name].data

    # Compute the number of unique parameter values
    n_param_values = len(grid_param)

    # Reshape the scores and standard deviations
    scores_mean = np.reshape(scores_mean, (-1, n_param_values))
    scores_sd = np.reshape(scores_sd, (-1, n_param_values))

    # Plot the scores for the parameter being optimised
    ax.plot(grid_param, scores_mean[0], '-o')
    ax.fill_between(grid_param, scores_mean[0] - scores_sd[0],
                    scores_mean[0] + scores_sd[0], alpha=0.1)

    # Add labels and titles to the plot
    ax.set_xlabel(param_name)
    ax.set_ylabel('CV Average Score')
    ax.set_title('Grid Search Scores for {}'.format(param_name))
```

### Multiple Parameter Results Plot

```python
def plot_grid_search_results(cv_results, params, figsize=(20, 5)):
    """
    Plot the grid search results for multiple parameters. Uses plot_grid_search_one_param for each parameter value. 

    Parameters:
        cv_results (dict): The dictionary of cross-validation results from grid search.
        params (dict): A dictionary mapping parameter names to their corresponding values.
        figsize (tuple, optional): The figure size (width, height) in inches. Default is (20, 5).

    Returns:
        None

    """
    display(Markdown("### Grid Search Results"))
    fig, axs = plt.subplots(1, len(params), figsize=figsize)

    # Iterate over each parameter and plot the grid search results
    for i, param_name in enumerate(params.keys()):
        plot_grid_search_one_param(
            cv_results, params[param_name], param_name, ax=axs[i])
    plt.tight_layout()
    plt.show()
```

# Evaluate the impact of Split Ratios on Performance
The following section will use the optimised versions of the models explored above in order to determine the affect of altering the split ration between the training and testing data on the performance of our regressor and classifier models respectively. 

## Split Ratio Evaluation Functions

```python
def evaluate_split_ratio(model, X, y, split_ratio=0.2):
    """
    Evaluates the impact of split ratio on performance using a machine learning model and dataset.
    
    Parameters:
        model (object): A machine learning model object with fit and predict methods.
        X (numpy.ndarray): The input features of the dataset.
        y (numpy.ndarray): The target variable of the dataset.
        split_ratio (float, optional): The ratio to split the dataset into training and testing sets.
                                       Defaults to 0.2 (20% for testing).
                                       
    Returns:
        float: The performance metric score of the model on the testing set.
    """
    # Split the dataset into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split_ratio, random_state=42)
    
    # Fit the model on the training data
    model.fit(X_train, y_train)
    
    # Predict using the trained model on the testing data
    y_pred = model.predict(X_test)
    
    # Evaluate the performance metric of the model on the testing set
    performance_score = model.score(X_test, y_test)
    
    return performance_score

def evaluate_range_of_splits(model, data, split_ratios):
    """
    Calls the evaluate_split_ratio function for a range of split ratios and evaluates the results.
    
    Parameters:
        model (object): A machine learning model object with fit and predict methods.
        X (array-like): The input features of the dataset.
        y (array-like): The target variable of the dataset.
        split_ratios (list): A list of split ratios to evaluate.
        
    Returns:
        dict: A dictionary containing the split ratios as keys and the corresponding performance scores as values.
    """
    X = data[:, :-1]
    y = data[:, -1]
    
    results = {}
    
    for split_ratio in split_ratios:
        score = evaluate_split_ratio(model, X, y, split_ratio)
        results[split_ratio] = score
        
    split_ratios = list(results.keys())
    scores = list(results.values())
    
    plt.plot(split_ratios, scores, marker='o')
    plt.xlabel('Split Ratio')
    plt.ylabel('Performance Score')
    plt.title('Impact of Split Ratio on Performance')
    plt.show()
    
    return results
```
